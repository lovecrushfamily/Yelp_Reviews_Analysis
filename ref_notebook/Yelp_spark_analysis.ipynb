## Nombre d’établissements référencés par Yelp dans chaque état

**Objectif** : obtenir un dataframe (df) avec le nombre d'établissements dans chaque état pour pouvoir ensuite créer une carte rendant compte de la distribution des restaurants dans les états américains.

Etape 1 : On regroupe les restaurants par états en utilisant le df tip qui contient une ligne pour chaque restaurant dans Yelp et un groupe by('state').

Etape 2 : On compte le nombre de lignes (chaque ligne correspondant à un restaurant) pour chaque états (suite au groupby)

## Villes ayant en moyenne les meilleurs restaurants

**Objectif** : Afficher les villes ayant en moyenne les meilleurs restaurants (beaucoup de reviews et bonnes notes).

Etape 1 : On regroupe les restaurants par villes en utilisant le df tip qui contient une ligne pour chaque restaurant dans Yelp et un groupe by('city').

Etape 2 : On filtre le df obtenu suite au groupby. Une ville sera considérée comme une des best city si elle a des restaurants avec en moyenne : plus de 50 reviews et une note supérieur à 4 (filtre que l'on réalise à l'aide de la méthode filter).

## Etats ayant le plus de commentaire contenant le mot "best"

Objectif : Obtenir les etats dans lesquels il y a le plus de reviews contenant le mot "best".

Etape 1 : Contrairement à précédemment, le df tip n'est plus sufffisant car il ne contient pas d'informations sur les reviews. On effectue donc un outer join (récupération de toutes les données) entre df_b et df_t pour obtenir les données nécessaires dans un même dataframe que nous appellons df_join.

```python
df_join = df_t.join(df_b,df_t.business_id ==  df_b.business_id,"outer")

df_join.show()
```

Etape 2 : Puis, on filtre sur les commentaires contenant le mot "best" avec la méthode filter et la méthode like.

```python
df_state_with_most_bestcom = df_join.filter(col('text').like("%best%"))
df_state_with_most_bestcom.describe().show()
```

Etape 3 : On effectue un groupby state et un count pour obtenir, comme on a pu le faire pour la première carte, le nombre de commentaire contenant le mot best dans chaque état. Cela nous donne un df constituer de deux colonnes qui seront utilisées pour effectuer la carte avec folium.

```python
df_state_with_most_bestcom = df_join.groupBy('state').count()
df_state_with_most_bestcom.show(15)
```

```python
df_state_with_most_bestcom_pandas = df_state_with_most_bestcom.toPandas()
```

## Machine Learning

Ici nous nous intéressons au sentiment d'une review en fonction du texte de celle-ci, on va donc créer un dataframe avec une colonne avec le texte de la review et une colone avec le sentiment de la review :

```python
drop_list = ['business_id', 'cool', 'date', 'funny', 'review_id', 'useful', 'user_id']
data_preprocessed = df_r.select([column for column in df_r.columns if column not in drop_list])
data_preprocessed.show(5)
```

```python
from pyspark.sql.functions import when, col, lit

data_preprocessed = data_preprocessed.withColumn('Feeling', when(col('stars') >= 3, lit(1)).otherwise(lit(0)))
data_preprocessed.show()
```

```python
data_preprocessed = data_preprocessed.select(['text', 'Feeling'])
data_preprocessed.show()
```

```python
from pyspark.ml.feature import RegexTokenizer, StopWordsRemover, CountVectorizer
from pyspark.ml import Pipeline

# regular expression tokenizer
regexTokenizer = RegexTokenizer(inputCol="text", outputCol="words", pattern="\\W")
# stop words
add_stopwords = ["http","https","amp","rt","t","c","the"] 
stopwordsRemover = StopWordsRemover(inputCol="words", outputCol="filtered").setStopWords(add_stopwords)
# bag of words count
countVectors = CountVectorizer(inputCol="filtered", outputCol="features", vocabSize=10000, minDF=5)

pipeline = Pipeline(stages=[regexTokenizer, stopwordsRemover, countVectors])
# Fit the pipeline to training documents.
pipelineFit = pipeline.fit(data_preprocessed)
dataset = pipelineFit.transform(data_preprocessed)
dataset.show(5)
```

```python
dataset = dataset.withColumnRenamed('Feeling', 'label')
dataset.show(5)
```

Maintenant on sépare nos données en train et test :

```python
(trainingData, testData) = dataset.randomSplit([0.7, 0.3], seed = 100)
```

```python
from pyspark.ml.classification import LogisticRegression

lr = LogisticRegression(maxIter=20, regParam=0.3, elasticNetParam=0)
lrModel = lr.fit(trainingData)
predictions = lrModel.transform(testData)
```
